## 7. Recursive Expression Evaluator

````python
def evaluate(expr):
    if isinstance(expr, int) or isinstance(expr, float):
        return expr

    op = expr[0]
    left_value = evaluate(expr[1])
    right_value = evaluate(expr[2])

    if op == "+":
        return left_value + right_value
    elif op == "*":
        return left_value * right_value
    else:
        raise ValueError(f"Unknown operator: {op}")

evaluate(["+", 1, ["*", 2, 3]])
````

````python
def evaluate_v2(expr):
    if isinstance(expr, int) or isinstance(expr, float):
        return expr

    op = expr[0]
    left_value = evaluate_v2(expr[1])
    right_value = evaluate_v2(expr[2])

    if op == "+":
        return left_value + right_value
    elif op == "*":
        return left_value * right_value
    elif op == "-":
        # TODO
        return left_value - right_value
    elif op == "/":
        # TODO
        return left_value / right_value
    else:
        raise ValueError(f"Unknown operator: {op}")

evaluate_v2(["-", ["*", 6, 2], 2])
````

## 8. Tiny REPL

````python
# Type expressions like:
# 1 + 2
# (1 + 2) * 3
# Type quit to stop.

while True:
    text = input(">>> ")

    if text == "quit":
        print("Goodbye!")
        break

    result = eval(text)
    print(result)
````
